```python
# Analysis of LBP data.
import re
import string
import time
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
import networkx as nx
import pyvis.network as net
from scipy.stats import entropy, pearsonr, f_oneway
import junctiontree as jt
```

```python
class Factor:
    dist = None

    def __init__(self, neigh, dist):
        self.variables = neigh
        self.dist = dist
        self.name = ", ".join([str(v) for v in self.variables])
        
    def __str__(self):
        return self.name
    
    def __lt__(self, other):
        return len(self.variables) < len(other.variables)
```

```python
class NodeFileParser:
    filepath = None
    rxs = \
        {
            "max": re.compile(r"MAX\s\d"),
            "nets": re.compile(r"NETS\s\d*"),
            "nodecount": re.compile(r"NODECOUNT\s\d\d\-\d\d"),
            "loops": re.compile(r"LOOPS\s\d*\-\d*")
        }

    def parseline(self, line: str) -> (str, str):
        """Apply regex to line iteratively"""
        for name, rx in self.rxs.items():
            match = rx.findall(line)
            if match:
                return name, match
        return None, None

    def parse(self, filepath: str) -> (int, int, list, list):
        """Parse nodefile specified at filepath
           Returns:
               max_loopsize (int): maximal loopsize
               max_nets (int): maximal networks per category
               nodecounts (int): number of nodes per network
               loopcounts (int): number of loops per network"""
        max_nets, max_loopsize, nodecounts, loopcounts = -1, -1, [], []

        # Read node file and retrieve data
        with open(filepath) as file:
            line = file.readline()
            while line:
                line.strip()
                name, contents = self.parseline(line)
                if contents:
                    content = contents[0]

                    # Retrieve maximal loopsize
                    if name == "max":
                        max_loopsize = int(content.split(" ")[1])
                        assert max_loopsize <= 4

                    # Retrieve maximal number of networks per category
                    elif name == "nets":
                        max_nets = int(content.split(" ")[1])

                    # Retrieve nodecounts
                    elif name == "nodecount":
                        ncs = content.split(" ")[1].split("-")
                        min_nc = int(ncs[0])
                        max_nc = int(ncs[1])
                        assert min_nc <= max_nc
                        nodecounts.append((min_nc, max_nc))

                    # Retrieve number of loops
                    elif name == "loops":
                        lcs = content.split(" ")[1].split("-")
                        min_lc = int(lcs[0])
                        max_lc = int(lcs[1])
                        assert min_lc <= max_lc
                        loopcounts.append((min_lc, max_lc))

                line = file.readline()
            file.close()
        return max_loopsize, max_nets, nodecounts, loopcounts


class NetworkFactory:
    settings = []
    set = False
    parser = NodeFileParser()

    @staticmethod
    def generate_distribution() -> np.ndarray:
        """Generate a probability distribution. 
        Distribution is later normalized over one of the axes
        for consistency."""
        return np.random.randint(1, 20, size=(2, 2))

    @staticmethod
    def random_name(length: int) -> str:
        """Generate a random name of some length"""
        letters = string.ascii_lowercase
        return ''.join(random.choice(letters) for i in range(length))

    def connect_random_factors(self, f1: Factor, f2: Factor) -> Factor:
        """Connect two factors by a random variable.
           Returns:
                a factor connecting the two given Factors
        """
        v1 = random.choice(f1.variables)
        v2 = random.choice(f2.variables)
        return Factor([v1, v2], self.generate_distribution())

    def generate_loop(self, looplength: int) -> list:
        """Generate a loop of particular length.
        Returns:
            loop (list of Factors): loop consisting of Factors"""

        chain = self.generate_chain(looplength)
        # Connect final and first vertex
        chain.append(Factor([chain[-1].variables[1],
                     chain[0].variables[0]],
                     self.generate_distribution()))

        # Chain is now a loop
        loop = chain
        return loop

    def generate_chain(self, length: int) -> list:
        """Generate chain of factors of particular length."""
        chain = []
        idlist = [self.random_name(4) for i in range(length)]
        # Generate chain of vertices
        for i in range(0, length-1):
            vertex = Factor([idlist[i], idlist[i+1]],
                            self.generate_distribution())
            chain.append(vertex)
        return chain

    def generate_settings(self, MAX_LOOPSIZE: int, max_networks: int,
                          nodecounts: list, loopcounts: list):
        """Generate consistent settings per network"""
        MIN_LOOPSIZE = 2
        
        def random_number(low: int, high: int) -> int:
            """Return random number between low and high bounds"""
            return random.choice(list(range(low, high)))
        
        network_settings = []
        
        # Monster nested loop, 
        # does not always generate settings due to constrains set on network generation.
        # If no result within 5 seconds, try again.
        for n in nodecounts:
            for loop in loopcounts:
                for net in range(max_networks):
                    loopcount = 10000    # 'Impossibly' high number
                    nr_nodes = random_number(n[0], n[1]+1)
                    loopsize = random_number(2, MAX_LOOPSIZE)
                    low_bound, high_bound = loop    # Bounds for number of loops
                    while loopsize * loopcount >= nr_nodes-loopsize or loopcount == 10000:
                        if low_bound * MIN_LOOPSIZE > nr_nodes:   # If number of nodes and number of loops 
                            low_bound =- 1
                        loopcount = random_number(low_bound, high_bound)
                        loopsize = random_number(MIN_LOOPSIZE, MAX_LOOPSIZE)

                    network_settings.append([loopcount,
                                             nr_nodes,
                                             loopsize])
                    
        self.write_settings(network_settings)  # Write settings to file

    def generate_network(self, loopcount: int,
                         number_nodes: int, loopsize: int) -> list:
        """ Generate network. 
        Params:
            loopcount (int): the number of loops in the network
            number_nodes (int): number of nodes in the network
            loopsize (int): length of each loop in the network
            """
        loops = []
        connections = []

        # Generate required number of loops
        for i in range(loopcount):
            for l in self.generate_loop(loopsize):
                loops.append(l)

        # Connect loops to each other
        for i in range(1, (loopcount*loopsize)):
            if i % loopsize == 0:
                connections.append(
                    self.connect_random_factors(loops[i], loops[i-1]))

        connected_loops = loops+connections

        # Connect remaining vertices to loops in a long chain, if possible
        chainlength = number_nodes-loopsize*loopcount
        if chainlength > 1:
            chain = self.generate_chain(chainlength)   # Create chain
            loop = random.choice(connected_loops)      # Choose loop
            vertex = random.choice(chain)              # Choose vertex
            conn = self.connect_random_factors(loop, vertex)
            chain.append(conn)                         # Connect loop and vertx
            connected_chain = chain                    # Chain is now connected
            return connected_loops + connected_chain

        return connected_loops

    def write_settings(self, settings):
        """Write settings to file"""
        with open("text_files/settings.txt", 'w') as f:
            f.write("netconfig\n")
            for setting in settings:
                for s in setting:
                    f.write(str(s)+" ")
                f.write("\n")
            f.close()

    def read_settings(self):
        """Read settings from file"""
        sets = []
        with open("text_files/settings.txt", "r") as f:
            lines = f.readlines()
            for line in lines[1:]:
                linedata = line.strip().split(" ")
                data = [int(d) for d in linedata]
                sets.append(data)
            f.close()
        return sets

    def create_settings(self, path_to_nf: str):
        """Parse requirements from file and generate networks accordingly
        Params:
            path_to_nf (str): path to file to read requirements from
        """
        max_loopsize, \
            max_networks, \
            nodecounts, \
            loopcounts = self.parser.parse(path_to_nf)
        self.generate_settings(max_loopsize, max_networks,
                               nodecounts, loopcounts)

    def make_networks(self, new=False):
        """Generate settings for each network."""
        if new:
            self.create_settings("text_files/node_file.txt")
        settings = self.read_settings()
        networks = []
        for s in settings:
            loopcount = s[0]
            number_nodes = s[1]
            loopsize = s[2]
            network = self.generate_network(loopcount, number_nodes, loopsize)
            networks.append([network, number_nodes, loopcount, loopsize])
        print(len(networks))
        return networks
```

```python
class LBP:
    def __init__(self, facs):
        self.facs = facs
        self.messages = {}
        self.vars = []
        self.set_vars()
        self.initialize_to_ones()
        self.ran = False
        self.itr = 0

    def set_vars(self):
        """Set variables from factors."""
        s = []
        for f in self.facs:
            for v in f.variables:
                s.append(v)
        self.vars = list(set(s))  # Remove duplicates

    def get_graph(self) -> (list, list):
        """Get vertices and edges from factors"""
        edges = []
        fs = []
        vs = []
        for f in self.facs:
            fs.append(f.name)
            for v in f.variables:
                if v not in vs:
                    vs.append(v)
                edges.append((f.name, v))
                edges.append((v, f.name))
        vertices = fs + vs
        return (vertices, edges)

    def initialize_to_ones(self):
        """Initialize messages to ones."""
        for f in self.facs:
            f.name = str(f.variables)
            for i, var in enumerate(f.variables):
                self.messages[(var, f.name)] = Factor([var],
                                                      np.ones(f.dist.shape[i]))
                self.messages[(f.name, var)] = Factor([var],
                                                      np.ones(f.dist.shape[i]))

    def arr_prod(self, arr: list) -> Factor:
        """Return product of list of Factors
        Args:
            arr (list of Factors): list over which to compute product
        Returns:
            result (Factor): product of arr"""
        result = arr[0]
        for a in arr[1:]:
            result = self.product(a, result)
        return result

    def product(self, f1: Factor, f2: Factor) -> Factor:
        """Compute product of two factors"""
        # Get intersecting variables of factors
        f12, f12_f1, f12_f2 = np.intersect1d(f1.variables,
                                             f2.variables, return_indices=True)

        # Get difference between variables of f1 and f2
        diff_12 = np.setdiff1d(f1.variables, f2.variables, assume_unique=True)
        diff_21 = np.setdiff1d(f2.variables, f1.variables, assume_unique=True)

        # Initialize indices of new distribution
        f1_ind = np.full(len(f1.variables), -1)
        f2_ind = np.full(len(f2.variables), -1)

        # Find variables of f1 and f2 not in combined factor
        f1_mask = np.isin(f1.variables, f12, invert=True)
        f2_mask = np.isin(f2.variables, f12, invert=True)

        # Get new ax positions
        f1_ind[f1_mask] = np.arange(np.sum(f1_mask))
        f2_ind[f2_mask] = np.arange(np.sum(f2_mask)) + \
            np.sum(np.invert(f2_mask))
        f1_ind[f12_f1] = np.arange(len(f12)) + np.sum(f1_mask)
        f2_ind[f12_f2] = np.arange(len(f12))

        # Move axes into new positions to ensure proper order
        f1_dist = np.moveaxis(f1.dist, range(len(f1_ind)), f1_ind)
        f2_dist = np.moveaxis(f2.dist, range(len(f2_ind)), f2_ind)

        # Compute actual product
        f1_vars = tuple([slice(None)]*len(f1.variables)+[None]*len(diff_21))
        f2_vars = tuple([None]*len(diff_12)+[slice(None)])
        prod = f1_dist[f1_vars] * f2_dist[f2_vars]
        return Factor(list(diff_12)+list(f12)+list(diff_21), prod)

    def marginalize(self, f: Factor, vars: list) -> Factor:
        """Marginalize factor according to specified variable(s)
        Args:
            f (Factor): Factor to marginalize
            vars (list of str): variables to marginalize out""" 
        diff = np.setdiff1d(f.variables, vars, assume_unique=True)
        ax = tuple(np.where(np.isin(f.variables, vars))[0])
        return Factor(diff, np.sum(f.dist, axis=ax))

    def normalize(self, arr: Factor) -> Factor:
        """Normalize factor
        Args:
            arr (Factor): Factor which to normalize
        Returns:
            normalized Factor"""
        return Factor(arr.variables, arr.dist/sum(arr.dist))

    def factor_to_variable(self, fac: Factor, var: str) -> Factor:
        """ Compute factor to variable message.
        Args:
            fac (Factor): sender of message, factor
            var (Variable): recipient of message, variable

        Returns:
            The product of the factor with messages from all its neighbors
            otherwise
        """
        messages = []
        messages.append(fac)
        margs = []
        for v in fac.variables:
            if v != var:
                message = self.get_var_to_fac(v, fac.name)
                messages.append(message)
                margs.append(v)
        return self.normalize(self.marginalize(self.arr_prod(messages), margs))

    def variable_to_factor(self, var: str, fac: Factor) -> Factor:
        """ Compute variable to factor message."""
        values = []
        neighbors = [f for f in self.facs if var in f.variables]
        for f in neighbors:
            if f != fac.name:
                message = self.get_fac_to_var(f.name, var)
                values.append(message)
        return self.normalize(self.arr_prod(values))

    def get_fac_to_var(self, fac: str, var: str) -> Factor:
        """Return factor to variable message"""
        return self.messages[(fac, var)]

    def get_var_to_fac(self, var: str, fac: str) -> Factor:
        """Return variable to factor message"""
        return self.messages[(var, fac)]

    def run(self):
        """Run LBP until convergence"""
        new_msg = {}
        d = False
        factors = {f.name: f for f in self.facs}
        iter = 0
        print("run")
        while not d:
            for e in self.messages.keys():
                fr, to = e[0], e[1]
                if fr in factors:
                    fac = factors[fr]
                    new_msg[(fac.name, to)] = self.factor_to_variable(fac, to)
                    new_msg[(to, fac.name)] = self.variable_to_factor(to, fac)
                else:
                    fac = factors[to]
                    new_msg[(fr, fac.name)] = self.variable_to_factor(fr, fac)
                    new_msg[(fac.name, fr)] = self.factor_to_variable(fac, fr)
            news = [n.dist for n in list(new_msg.values())]
            olds = [m.dist for m in list(self.messages.values())]
            d = np.allclose(olds, news) or iter > 1000   #Elementwise equality within very small tolerance
            self.messages.update(new_msg)
            iter += 1
        self.ran = True
        return iter

    def get_var_dist(self, var_name: str) -> np.ndarray, int:
        """Get distribution of query variable
        Args:
            var_name (str): name of query variable
        Returns: 
            (ndarray, int): (normalized marginal distribution, number of iterations)
            """
        
        if not self.ran:
            self.itr = self.run()
        messages = []
        for f in self.facs:
            if var_name in f.variables:
                messages.append(self.get_fac_to_var(f.name, var_name))
        dst = self.arr_prod(messages).dist
        return dst/sum(dst), self.itr
```

```python
print(len(loops[-1]))
nws = [i for i in [n[0] for n in networks]]
true_distributions = []
for i in range(len(nws)):
    jt_lbp = LBP(nws[i])
    vrs = []
    for n in nws[i]:
        vrs += n.variables
    key_sizes = {v: 2 for v in vrs}
    factors = [n.variables for n in nws[i]]
    values = [n.dist/np.sum(n.dist, axis=1) for n in nws[i]]
    tree = jt.create_junction_tree(factors, key_sizes)
    prop_values = tree.propagate(values)
    sums = []
    for n in range(10):
        s = np.sum(prop_values[n], axis=1)
        sums.append(s/np.sum(s))
    true_distributions.append(sums)
 
```

## Analysis by loop count

```python
def compute_entropy(df):
    ent = []
    for i in range(len(df)):
        ds = [df["d0"].to_numpy()[i], df["d1"].to_numpy()[i]]
        ts = [df["t0"].to_numpy()[i], df["t1"].to_numpy()[i]]
        ent.append(entropy(ds, ts))
    return ent

def analyze(df):
    df = df.reset_index(drop=True)
    print("Correlation", pearsonr(df["d1"], df["t1"]))
    itr = df["iter"].reset_index(drop=True)
    plt.scatter(df["d1"], df["t1"])
    plt.show()
    plt.boxplot(df["accuracy"])
    plt.show()
    plt.plot(df["d1"]/df["t1"])
    plt.show()
    plt.plot(compute_entropy(df[0:36]))
    print("Average KL-divergence: ", np.average(compute_entropy(df[0:36])))
    print("Average accuracy: ", np.average(df["accuracy"]))
    plt.show()
    plt.title("iterations")
    plt.plot(itr)
    print("Average computation time: ", np.average(df["time"]))
    print("Average iterations: ", np.average(itr))
    
def reject_outliers(data, m=2):
    return data[abs(data - np.mean(data)) < m * np.std(data)]
analyze(data)
```
